### Trimming Data

```python
actual = Actual_Demand[["Segment", "Product", "Mean"]].copy()

actual = actual.rename(columns={"Mean": "mean_demand"})

actual
```

```python
two_m = Two_month_fc[["Segment", "Product", "sqrt(MSE)"]].copy()
two_m = two_m.rename(columns={"sqrt(MSE)": "2mo_sqrt(MSE)"})
two_m
```

```python
six_m = Six_month_fc[["Segment", "Product", "sqrt(MSE)"]].copy()
six_m = six_m.rename(columns={"sqrt(MSE)": "6mo_sqrt(MSE)"})
six_m
```

```python
# Megre into a single table

df = actual.merge(two_m, on=["Segment", "Product"], how="inner")
df = df.merge(six_m, on=["Segment", "Product"], how="inner")

df
```

### Newsvendor model

```python
# Defining underage and overage costs and critical ratio

underage_cost = 0.20   # profit from selling one extra unit = 20% of wholesale price
overage_cost = 0.065   # loss from one extra unsold unit = 6.5% of wholesale price

critical_fractile = underage_cost / (underage_cost + overage_cost)
z_value = norm.ppf(critical_fractile)

print("Critical fractile:", critical_fractile)
print("Z-value:", z_value)
```

```python
df["Q_star_2m"] = df["mean_demand"] + z_value * df["2mo_sqrt(MSE)"]
df["Q_star_6m"] = df["mean_demand"] + z_value * df["6mo_sqrt(MSE)"]

# Since MRI machines are whole units, round up to the next integer
df["Q_star_2m"] = np.ceil(df["Q_star_2m"]).astype(int)
df["Q_star_6m"] = np.ceil(df["Q_star_6m"]).astype(int)

print(df[["Segment", "Product", "mean_demand", "2mo_sqrt(MSE)", "6mo_sqrt(MSE)", "Q_star_2m", "Q_star_6m"]])
```

### Calculating profits

```python
# Keeping only the price columns we need
price = Price[["Segment", "Product", "Unit"]].copy()

# Merging prices with quantity data
df = df.merge(price, on=["Segment", "Product"], how="inner")

# Check result
print(df[["Segment", "Product", "mean_demand", "Q_star_2m", "Q_star_6m", "2mo_sqrt(MSE)", "6mo_sqrt(MSE)", "Unit"]])
```

```python
# Definining underage and overage costs as actual costs (instead of a % of price)

underage_rate = 0.20   # profit on one sold unit
overage_rate = 0.065   # loss on one leftover unit

df["profit_per_sold_unit"] = underage_rate * df["Unit"]
df["loss_per_leftover_unit"] = overage_rate * df["Unit"]

print(df[["Segment", "Product", "Unit", "profit_per_sold_unit", "loss_per_leftover_unit"]])
```

```python
# Calculating sold and leftover units

# 2-month scenario
df["expected_sold_2m"] = np.minimum(df["Q_star_2m"], df["mean_demand"])
df["expected_leftover_2m"] = np.maximum(df["Q_star_2m"] - df["mean_demand"], 0)

# 6-month scenario
df["expected_sold_6m"] = np.minimum(df["Q_star_6m"], df["mean_demand"])
df["expected_leftover_6m"] = np.maximum(df["Q_star_6m"] - df["mean_demand"], 0)

print(df[[
    "Segment", "Product", "mean_demand",
    "Q_star_2m", "expected_sold_2m", "expected_leftover_2m",
    "Q_star_6m", "expected_sold_6m", "expected_leftover_6m"
]])
```

```python
#Calculating expected profits

# 2-month expected profit
df["expected_profit_2m"] = (
    df["expected_sold_2m"] * df["profit_per_sold_unit"]
    - df["expected_leftover_2m"] * df["loss_per_leftover_unit"]
)

# 6-month expected profit
df["expected_profit_6m"] = (
    df["expected_sold_6m"] * df["profit_per_sold_unit"]
    - df["expected_leftover_6m"] * df["loss_per_leftover_unit"]
)

df[[
    "Segment", "Product",
    "expected_profit_2m", "expected_profit_6m"
]]
```

```python
# Compare twwo months vs 6 months

df["profit_difference"] = df["expected_profit_2m"] - df["expected_profit_6m"]
df["Q_difference"] = df["Q_star_6m"] - df["Q_star_2m"]

df["better_scenario"] = np.where(
    df["expected_profit_2m"] > df["expected_profit_6m"],
    "2-month",
    "6-month"
)

comparison = df[[
    "Segment",
    "Product",
    "mean_demand",
    "2mo_sqrt(MSE)",
    "6mo_sqrt(MSE)",
    "Q_star_2m",
    "Q_star_6m",
    "expected_profit_2m",
    "expected_profit_6m",
    "profit_difference",
    "Q_difference",
    "better_scenario"
]]

comparison
```

```python
# Rounding for readability
comparison_rounded = comparison.copy()

comparison_rounded["expected_profit_2m"] = comparison_rounded["expected_profit_2m"].round(2)
comparison_rounded["expected_profit_6m"] = comparison_rounded["expected_profit_6m"].round(2)
comparison_rounded["profit_difference"] = comparison_rounded["profit_difference"].round(2)
comparison_rounded["mean_demand"] = comparison_rounded["mean_demand"].round(2)
comparison_rounded["2mo_sqrt(MSE)"] = comparison_rounded["2mo_sqrt(MSE)"].round(2)
comparison_rounded["6mo_sqrt(MSE)"] = comparison_rounded["6mo_sqrt(MSE)"].round(2)

comparison_rounded
```
